```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from functools import reduce
import scipy
from scipy import sparse
from scipy.special import gamma
import scipy.sparse.linalg
from scipy.integrate import cumulative_trapezoid
from fractions import Fraction
```

```python
def f(x):
    return np.exp(-x) / (1+(x-1)**2)
def Int_est(a,b,N):
    samp = np.random.uniform(a,b,N)
    f_val = f(samp)
    I = np.mean(f_val)*(b-a)
    return I
def error(fun,bin,**kwargs):
    Ig=[]
    for i in range(bin):
        Ig.append(fun(**kwargs))

    return np.std(Ig)
    
    
a=0
b=10
Num = 1000000
kwargs = {'a':a,'b':b,'N':Num}
error_est= error(Int_est,50,**kwargs)
Integral = Int_est(a,b,Num)
print(f'{Integral: .5f}')
print(f'{error_est: .5f}')
```

$$
 {pdf \Rightarrow g(x) = \alpha * e^{-\alpha x}}
$$
$$
G(y) = {\int_0^x{pdf(x)}\,dx}  =  -e^{-\alpha x} +1
$$
 $$ x_{sampled}= G^- (y) =-\ln{(1-y)}/\alpha $$

 $$
 \int_a^b{f(x) \,dx} = \int_0^1 {\frac{f(G^-(y))}{g(G^-(y))}}
 $$

```python
def f(x):
    return np.exp(-x) / (1+(x-1)**2)
def pdf(x,alp):
    return alp*(np.exp(-alp*x))
def invG(y,alp):
    return -np.log(np.abs(1-y))/alp
def sampI(N):
    y=np.random.uniform(0,1,N)
    y=np.clip(y,1e-10,1-1e-10)
    samp_est = (f(invG(y,alp))/pdf(invG(y,alp), alp))
    return np.mean(samp_est)

```

```python
alp = 1.46
sampI_list= []
samperr_list = []
N_list = np.random.randint(1000,100000, size = 500)
for j in N_list:
    
    arge ={'N':j}
    si = sampI(j)
    se = error(sampI,50,**arge)
    sampI_list.append(si)
    samperr_list.append(se)
I_list= []
err_list = []
#N_list = np.random.randint(100,100000, size = 5000)
for k in N_list:
    ars={'a':a,'b':b,'N':k}
    i = Int_est(a,b,k)
    e= error(Int_est,50,**ars)
    I_list.append(i)
    err_list.append(e)
```

```python
sample_size = int(1 * len(N_list))
indices = np.random.choice(len(N_list), sample_size)

print(f' estimation W imp. sampling= {(np.mean(sampI_list)): .5f}, error = {np.mean(samperr_list): .5f}')
print(f'estimation W/O imp. sampling= {np.mean(I_list): .5f}, error = {np.mean(err_list): .5f}')

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.scatter(N_list,sampI_list, marker='o')
#plt.scatter(indices,[I_list[t]for t in indices], marker='o')
plt.ylabel('est W Imp_samp')
plt.xlabel('number of samples')
plt.axhline(np.mean(sampI_list),linestyle='--',linewidth=2.5, c= 'Red', label= 'average estimate of Integral' )

plt.legend()

#plt.errorbar(indices, [I_list[t] for t in indices], capsize=1,yerr= 0.5*np.std(I_list),fmt='.', c='green')

#plt.errorbar(N_list, I_list, capsize=1,yerr= np.std(I_list),fmt='.', c='green')

plt.subplot(2,2,2)
plt.scatter(N_list,samperr_list, marker='.')
plt.xlabel('number of samples')
plt.ylabel('error W Imp_samp')
plt.xscale('log')
plt.yscale('log')
plt.plot(N_list,1/(np.sqrt(N_list)), c='r', linestyle='--',label='MC error ')
#plt.axhline(np.mean(samperr_list), c='r', linestyle='--',label='mean error ')
plt.legend()



plt.subplot(2,2,3)
plt.scatter(N_list,I_list, marker='o')
#plt.scatter(indices,[I_list[t]for t in indices], marker='o')
plt.ylabel('est WO Imp_samp')
plt.xlabel('number of samples')
plt.axhline(np.mean(I_list),linestyle='--',linewidth=2.5, c= 'Red', label= 'average estimate of Integral' )
plt.legend()


plt.subplot(2,2,4)
plt.scatter(N_list,err_list, marker='.')
plt.xlabel('number of samples')
plt.ylabel('error WO Imp_samp')
plt.xscale('log')
plt.yscale('log')

plt.plot(N_list,1/(np.sqrt(N_list)), c='r', linestyle='--',label='MC error ')
plt.legend()




plt.tight_layout()
plt.show()
```
